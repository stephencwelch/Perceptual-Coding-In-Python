```python
from PQEval import PQEval
PQE = PQEval(Amax = 32768)
```

```python
from scikits.audiolab import wavread

sigR, Fs, enc = wavread('exports/timeAligned_mic1_10sec.wav')
sigT, Fs, enc = wavread('exports/timeAligned_piezo_10sec.wav')
```

```python
NF = 2048
Nadv = NF / 2
Amax = 32768

#Number of critical bands:
Nc = 109
```

```python
#Number of frames:
Np = floor(len(sigR)/Nadv)
```

```python
#Scale up audio:
sigRS = Amax*sigR
sigTS = Amax*sigT
```

```python
#For testing purposes:
Np = 100
```

```python
#Create empty matrices:
X2 = np.zeros((2,NF/2+1))

X2MatR = np.zeros((Np, NF/2+1))
X2MatT = np.zeros((Np, NF/2+1))

EbNMat = np.zeros((Np, Nc))
EsMatR = np.zeros((Np, Nc))
EsMatT = np.zeros((Np, Nc))

EhsR = np.zeros((Np, Nc))
EhsT = np.zeros((Np, Nc))

previousFrameR = np.zeros(Nc)
previousFrameT = np.zeros(Nc)
```

```python
startS = 0
for i in range(Np):
    xR = sigRS[startS:NF+startS]
    xT = sigTS[startS:NF+startS]
    startS = startS+Nadv
    
    #Process Frame: 
    X2[0,:] = PQE.PQDFTFrame(xR)
    X2[1,:] = PQE.PQDFTFrame(xT)
    
    X2MatR[i,:] = X2[0,:]
    X2MatT[i,:] = X2[1,:]
    
    # Critical band grouping and frequency spreading
    EbN, Es = PQE.PQ_excitCB(X2)
    
    EbNMat[i,:] = EbN
    EsMatR[i,:] = Es[0,:]
    EsMatT[i,:] = Es[1,:]
    
    #Time domain spreading
    #Can't time-spread on first frame:
    if i > 0:
        EhsR[i,:], previousFrameR = PQE.PQ_timeSpread(EsMatR[i,:], previousFrameR)
        EhsT[i,:], previousFrameT = PQE.PQ_timeSpread(EsMatT[i,:], previousFrameT)
    else:
        #First Iteration:
        previousFrameR = EsMatR[i,:]
        previousFrameT = EsMatT[i,:]
        
```
