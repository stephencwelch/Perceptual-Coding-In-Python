##Noise to Mask Ratio (NMR)

Ok, NMR still looks pretty freaking good - I'll look into this one first!

So, NMR depends on 2 computed signals:

| Math Symbol| Variable Name in Code | Explanation | Dimension |
| :---------: | :-: | :-: | :-: |
| $\tilde{E}_{sR}[k, n]$ | EhsR | Frequency and Time spread excitation pattern for reference signal | numWindows x numCriticalBands |
| $ E_{bN}[k, n] $ | EbNMat | Difference of the magnitude squared signal, post frequency grouping| numWindows x numCriticalBands |

where $k$ is the frequency index and $n$ is our time (window) index.

According to Kabal's code, the difference of the magnitude signal is computed as:

$$
X_{wN}^2 = \mid X_{wR}\mid^2 - 2\sqrt{\mid X_{wR} \mid^2 \mid X_{wT} \mid^2} + \mid X_{wT} \mid^2 \tag{115}
$$

Where $X_{wR}^2$ and $X_{wT}^2$ are the (ear) weighted DFT magnitudes of the reference and test singals.

He does this computation in PQEval.m, lines 70-74:

```MATLAB
% Form the difference magnitude signal
for (k = 0:NF/2)
    XwN2(k+1) = (Xw2(1,k+1) - 2 * sqrt (Xw2(1,k+1) * Xw2(2,k+1)) ...
               + Xw2(2,k+1));
end
```

I don't understand why Kabal computes the difference this way - here's what the PEAQ stardard says to do:

$$
F_{Noise}[K_f, n] = \big\vert \vert F_{eref}[k_f,n] \vert - \vert F_{etest}[k_f,n] \vert \big\vert
$$

It says to take the difference of the absolute value of the magnitude signals. Notably, not the difference of the magnitude square signals.

Ok, so does kabal's implementation even make sense then?

Another twist here is that the difference signal, $X_{wN}^2$, is then frequency grouped to form $ E_{bN}[k, n]$.

Ok, so this is interesting, according to the standard, frequency spreading is applied to the squared-magnitude signal. Ok, so I think this is going to work then...

$$
X_{wN}^2 = (X_{wR} - X_{wT})^2 = X_{wR}^2 -2 X_{wR} X_{wT} + X_{wR}^2 \\
X_{wN}^2 = X_{wR}^2 - 2\sqrt{X_{wR}^2 X_{wT}^2} + X_{wT}^2
$$

Ok, so that's got to be what Kabal is doing - doing it this way is a bit easier to code, and eliminated the need for the absolute values in the stardard. The only loose end here is that we are not subrating the complex responses, instead the magnitude responses - I don't know how much this matters though. It would yield a difference result:

$$
X^2 = \Re^2 + \Im^2
$$

So given two complex signals, $X_1 = \Re_1 + \Im_1$ and $X_2 = \Re_2 + \Im_2$, 

$$
X_2 - X_1 = (\Re_2-\Re_1) + (\Im_2-\Im_1)
$$

The squared magnitude of the difference, $(X_2 - X_1)^2$, is then:

$$
(X_2 - X_1)^2 = (\Re_2-\Re_1)^2 + (\Im_2-\Im_1)^2 \\
= \Re_2^2 +\Re_2\Re_1 + \Re_1^2 +  \Im_2^2 +\Im_2\Im_1 + \Im_1^2
$$

###Clarification

So Kabal's implementation is different, but should yield the same results as the PEAQ stardard. The PEAQ standard calls for taking the absoluted value of the difference of the absolute values of the magnitude signals (In Kabal's notation):

$$
X_{wN} = \big\vert \vert X_{wR} \vert - \vert X_{wT} \vert \big\vert
$$

What we end up using in the next step (spreading) is $X_{wN}^2$, which may be computed as $X_{wN}^2 = (\vert X_{wR} \vert - \vert X_{wT} \vert)^2$ or as $X_{wN}^2 = X_{wR}^2 - 2\sqrt{X_{wR}^2 X_{wT}^2} + X_{wT}^2$. Kabal chooses to use the second option.

##Masking

The masking threshold is calculated form the excitation patterns. The parameter $m_{dB}[k]$ gives the amount by which the masking threshold lies below the time-frequency spread Bark energy. The parameter $m_{dB}[k]$ is a piecewise linear function on the bark scale.

This masking implementation seems pretty simple, and is definitely super relevant - worth definitely worth exploring.

````python
def PQ_MaskOffset(dz, Nc):
    gm = np.zeros(Nc)
    for k in xrange(Nc):
        if (k <= 12./dz):
            mdB = 3
        else:
            mdB = 0.25*k*dz  
        gm[k] = 10**(-1*float(mdB)/10) 
    return gm
````

````python
Nc, fc, fl, fu, dz = PQ.PQE.PQCB()
````

````python
gm = PQ_MaskOffset(dz, Nc)
````

````python
plot(gm)
grid(1)
````

Ok, so this is the "masking offset" (in energy units?) I think the interpretation here is that higher frequency sounds more difficult to mask.

The masking threshold is then: 

````python
M = PQ.EhsR*gm
````
